##EXPLORATORY DATA ANALYSIS OF OLYMPIC GAMES

Import library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

##Loading The Data Into The Dataframe

Loading the data into the pandas data frame is certainly one of the most important steps in EDA, as we can see that the value from the data set is comma-separated. So all we have to do is to just read the CSV into a data frame and pandas data frame does the job for us.

```python
df=pd.read_csv('/content/drive/MyDrive/project/olympic_games.csv')
df
# To display the top 5 rows
df.head(5)
```

```python
# To display bottom 5 rows
df.tail(5)
```

##1.Basic information about data-EDA

The df.info() function will give us the basic information about the dataset.For any data,it is good to start by knowing its information.Let's see how it work with our data.

```python
# Basic Information
df.info()
```

```python
# Describe the data
df.describe()
```

```python
# Length of data
len(df)
```

```python
# Shape of the data
df.shape
```

##2.Rename the columns

In this instance,most of the column names are very confusing to read,so i just tweakend their column names.This is a good approach it improves the readability of the data set.

```python
for i in df:
    j=i.replace(' ','_')
    df.rename(columns={i:j.capitalize()},inplace=True)
df.head()
```

##3.Unique values in the data

You can find the number of unique values in the particular column using unique() function in python

```python
df['Year'].unique()
df['Games_type'].unique()
df['Host_country'].unique()
df['Host_city'].unique()
df['Athletes'].unique()
df['Teams'].unique()
df['Country'].unique()
```

##4.Visualize The Unique Counts

Yes,you can visualize the unique values present in the data.For this we will be using the seaborn library.So you have to call the sns.countplot() function and specify the variables to plot the countplot.

```python
sns.countplot(df['Year'],color='red')
```

statistical summary gives a high level idea to identify whether the data has any outliers,data entry error,distribution of data such as the data is normally distributed or left/right skewed.

##7.Finding The Null Values

Finding the null values is the most important step in the EDA.As i told many a time,ensuring the quality of data is paramount.So let's see how we can find the null values.

```python
df.isnull().sum()
```

So here we can see that thier is no null values.Hence we can jump on to the next step in EDA.

##8.Duplicate Values

You can use the duplicate.sum() function to the sum of duplicate values present if any.It will show the number of duplicate values if they are present in the data.

```python
df.duplicated().sum()
```

So here we can see that their is no duplicate values present in the data. Therefore we enter into the next process in EDA.

##9.Statistical Function

You can use the df.mean() function to see the mean of the presented data.

```python
df.mean()
```

You can use the df.median() function to see the median of the presented data.

```python
df.median()
```

You can use the df.std() function to see the standard deviation of the presented data.

```python
df.std()
```

```python
a=df[(df['Teams'] > 90) & (df['Teams'] < 150)]
a.mean()
```

```python
b=df[(df['Athletes'] > 3000) & (df['Athletes'] < 10000)]
b.median()
```

```python
c=df[(df['Games_type'] == 'Summer')]
c.std()
```

##10.Filter The Data

Yes,you can filter the data based on some logic.

```python
df['Year'].head(10)
df['Athletes'].tail(10)
```

##11.Comparison Operator

You can combine multiple criteria and query your dataset as well. To do this, be sure to put each one in parentheses and use the logical operators | and & to separate them.

```python
df[(df['Gold'] == 0)]
```

```python
df[(df['Teams'] > 90) & (df['Teams'] < 150)]
```

```python
df[(df['Athletes'] > 3000) & (df['Athletes'] < 10000)]
```

##12.Correlation Plot - EDA

To find the correlation among the variables,we can make use of the correlation function.This will give you a fair idea of the correlation strength between different variables.

```python
df.corr()
```

##Heat Maps

Heat Maps is a type of plot which is necessary when we need to find the dependent variables. One of the best way to find the relationship between the features can be done using heat maps. In the below heat map we know that the Competition feature depends mainly on the Teams, Gold, Silver, Bronze.

```python
sns.heatmap(df.corr(),vmin=-1,vmax=1)
plt.show()
```

##Histogram

Histogram refers to the frequency of occurrence of variables in an interval..

```python
fig, axarr = plt.subplots(1,1, figsize=(12, 18))
df.groupby('Year')['Teams'].mean().sort_values(ascending=False).plot.bar( fontsize=12).set_title("Year And Teams", fontsize=18)
plt.subplots_adjust(hspace=1.0)
plt.subplots_adjust(wspace=.5)
sns.despine()
```

##Piechart

A piechart is another common EDA plot which is a circular diagram, often used in descriptive statistics for graphical representations of quantitative variables measured on classes of categories.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Teams", "Country", "Athletes", "Year"]
plt.pie(y, labels = mylabels)
plt.show()
```

##Box Plot

A box plot is another common EDA plot,which shows the distribution of a variable and can reveal outliers.

```python
df[['Teams']].boxplot()
```

```python
df[['Year']].boxplot()
```

##Line Plot

Line plot is the simplest plot, which displays data as a series of data points connected by straight line segments. A line plot is usually applied to time-series data and used to visualize the trend in the data.

```python
def plot():
  sns.lineplot(x=df['Year'],y=df['Teams'],data=df.head())
sns.set_palette('vlag')
plt.subplot(211)
plot()



def plot():
  sns.lineplot(x=df['Country'],y=df['Athletes'],data=df.head())
sns.set_palette('Accent')
plt.subplot(212)
plot()
```

##Scatter Plot

We generally use scatter plots to find the correlation between two variables. Here the scatter plots are plotted between Horsepower and Price and we can see the plot below. With the plot given below, we can easily draw a trend line. These features provide a good scattering of points.

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['Year'], df['Athletes'])
ax.set_xlabel('Year')
ax.set_ylabel('Athletes')
plt.show()
```
